Dari perhitungan ```Electric Range```, terlihat bahwa nilai rata-ratanya adalah ```61.5089928057554```
dimana angka ini sangat jauh dari nilai median nya yaitu ```0.0``` dan nilai standar deviasi yang sangat besar yaitu ```93.2717469427651```
yang menandakan bahwa besar kemungkinan adanya indikasi outlier. Untuk distribusi data nilai ```skew > 1``` yaitu ```positive skew/right skew``` dan
```nilai kurtosis > 0``` yaitu ```kurtosis leptokurtik/heavy tail```

### Mencari tahu berapa persentase outliers

```Karena skewness > |±0.5|, maka kita akan menggunakan tukey's rule untuk mencari outliers```

````python
# sort data first
df = df.sort_values(by=['Electric Range'])

# Calculating the IQR of a Pandas Column
quartiles = df['Electric Range'].quantile([0.25, 0.75])
q1 = quartiles[0.25]
q3 = quartiles[0.75]
iqr = q3-q1

# calculate upperBound and lowerBound outliers
upperBound = q3 + (1.5 * iqr)
lowerBound = q1 - (1.5 * iqr)

print("upperBound =", upperBound)
print("lowerBound =", lowerBound)
print("IQR =", iqr)
````

````python
# filter upperBound and lowerBound data
outliers = df.loc[(df['Electric Range'] < lowerBound) | (df['Electric Range'] > upperBound)]

# find length from each filters
totalOutlier = len(outliers)
totalData = len(df)

# calculate outlier percentage
outlierPercentage = totalOutlier/totalData*100

print("Total outlier =", totalOutlier)
print("Total data =", totalData)
print("Persentase outlier {}%".format(outlierPercentage))
````

## 2. Statistik Inferential

### Melakukan Uji hipotesis
1. Apakah ada perbedaan rata-rata antara jarak tempuh BEV dan PHEV?
- H0 = Rata-rata Electric Range antara kendaraan BEV dan PHEV adalah sama (µ1 == µ2)
- H1 = Rata-rata Electric Range kendaraan BEV lebih tinggi dibandingkan PHEV (µ1 != µ2)

````python
# Import modul statistik dari scipy untuk uji t-test
from scipy import stats
from scipy.stats import ttest_ind
````

````python
# filter data BEV dan PHEV dengan Electric Range > 0
bev_range = df[(df['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)') & (df['Electric Range'] > 0)]['Electric Range']
phev_range = df[(df['Electric Vehicle Type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)') & (df['Electric Range'] > 0)]['Electric Range']
````

```
Melakukan uji independent t-test untuk membandingkan rata-rata antara kendaraan BEV dan PHEV
```

````python
t_stat, p_value = ttest_ind(bev_range, phev_range, equal_var=False)

print("Statistik t:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Tolak H0: Terdapat perbedaan signifikan antara rata-rata Electric Range BEV dan PHEV")
else:
    print("Gagal tolak H0: Tidak terdapat perbedaan signifikan antara rata-rata Electric Range BEV dan PHEV")
````

Berdasarkan pengujian di atas, dapat disimpulkan bahwa:
- p-value = 0.0
- Critical Value = 0.05 (standar umum)
- Dikarenakan `p-value < critical value` maka H0 ditolak
- Dapat disimpulkan bahwa terdapat perbedaan signifikan rata-rata Electric Range antara BEV dan PHEV

### Melakukan uji hipotesis
2. Apakah jarak tempuh mempengaruhi penjualan electric Vehicle?
- H0 = Tidak ada hubungan antara rata-rata electric range dan jumlah penjualan kendaraan listrik per merek
- H1 = Ada hubungan antara rata-rata electric range dan jumlah penjualan kendaraan listrik per merek

````python
# Rata-rata Electric Range & total penjualan per merek (hanya yg > 0)
summary = df[df['Electric Range'] > 0].groupby('Make').agg({
    'Electric Range': 'mean',
    'City': 'count'  # jumlah data bisa dianggap proxy penjualan
}).rename(columns={'Electric Range': 'Average Range', 'City': 'Total Sales'}).reset_index()
````

```Untuk melihat seberapa kuat hubungan linear antara jarak tempuh dan penjualan, kita dapat menggunakan korelasi pearson```

````python
from scipy.stats import pearsonr

corr, p_val = pearsonr(summary['Average Range'], summary['Total Sales'])
print(f'Korelasi: {corr:.2f}, P-value: {p_val:.4f}')
````

Berdasarkan pengujian di atas, dapat disimpulkan bahwa:
- p-value = 0.0056
- Critical Value = 0.05 (standar umum)
- Dikarenakan `p-value < critical value` maka H0 ditolak
- Dapat disimpulkan bahwa terdapat hubungan antara rata-rata electric range dan jumlah penjualan kendaraan listrik per merek


````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Average Range', y='Total Sales', data=summary)

plt.title('Apakah Jarak Tempuh Mempengaruhi Penjualan EV?')
plt.xlabel('Rata-rata Jarak Tempuh')
plt.ylabel('Jumlah Penjualan (Proxy)')
plt.grid(True)
plt.show()
````

```
Dapat dilihat dari gambar visualisasi di atas bahwa semakin tinggi rata-rata jarak tempuh maka semakin tinggi jumlah penjualannya.
Hal ini menunjukkan adanya hubungan positif yang signifikan antara jarak tempuh kendaraan dan penjualan, yang berarti bahwa kendaraan 
listrik dengan rata-rata jarak tempuh lebih tinggi cenderung memiliki jumlah penjualan yang lebih tinggi.
```
